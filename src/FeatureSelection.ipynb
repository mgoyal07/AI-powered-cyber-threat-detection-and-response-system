```python
import os
import numpy as np
import pandas as pd
import cupy as cp
from cuml.ensemble import RandomForestClassifier
from sklearn.metrics import log_loss
import matplotlib.pyplot as plt

# List of attack types and BENIGN
# attack_types = [
#     "Bot",
#     "DDoS",
#     "DoS GoldenEye",
#     "DoS Hulk",
#     "DoS Slowhttptest",
#     "DoS slowloris",
#     "FTP-Patator",
#     "Heartbleed",
#     "Infiltration",
#     "PortScan",
#     "SSH-Patator",
#     "Web Attack - Brute Force",
#     "Web Attack - Sql Injection",
#     "Web Attack - XSS"
# ]
attack_types = ["Bot", "DDoS", "DoS GoldenEye", "DoS Hulk", "DoS Slowhttptest", "DoS slowloris", "FTP-Patator",
                "PortScan", "SSH-Patator"]
benign_type = "BENIGN"

def perform_feature_selection_single_feature(data):
    """
    For each feature in 'data' (except ' Label'), trains a GPU-based RandomForestClassifier
    to predict the label (benign=0, attack=1) using ONLY that single feature.

    - Computes log_loss for each single-feature model (lower is better).
    - Converts log_loss to 'importance' by importance = 1 / (log_loss + epsilon).
    - Returns a 1D NumPy array of these importance scores, in the same order as the features.
    """
    # Columns except " Label" are our features
    feature_cols = data.drop(columns=[" Label"]).columns

    # Convert the label column to 0=benign,1=attack on CPU first
    y_cpu = data[" Label"].apply(lambda x: 1 if x != benign_type else 0).values

    # We'll store an importance score for each feature
    importances_list = []

    # A small epsilon to avoid dividing by zero if log_loss=0
    epsilon = 1e-12

    for feat in feature_cols:
        # 1) Extract this single feature as X (CPU -> GPU)
        X_col_cpu = data[[feat]].values  # shape (n_samples, 1)
        X_col_gpu = cp.asarray(X_col_cpu)
        y_gpu = cp.asarray(y_cpu)

        # 2) Train a single-feature random forest
        clf = RandomForestClassifier(
            n_estimators=100,
            max_features=1.0,
            random_state=42,
            n_streams=8
        )
        clf.fit(X_col_gpu, y_gpu)

        # 3) Predict probabilities and compute log_loss on CPU
        prob_gpu = clf.predict_proba(X_col_gpu)   # shape (n_samples, 2)
        prob_cpu = prob_gpu.get()                # Cupy -> NumPy
        # y_cpu is already a NumPy array
        ll = log_loss(y_cpu, prob_cpu)

        # 4) Convert log_loss to "importance" (larger => more “important”)
        #    We'll do: importance = 1 / (log_loss + epsilon).
        #    If a feature yields lower log_loss, it gets higher importance.
        imp = 1.0 / (ll + epsilon)
        importances_list.append(imp)

    # Convert to NumPy array
    importances_array = np.array(importances_list)
    return importances_array


########################################
# Main loop over your attack CSVs
########################################
for attack_type in attack_types:
    # Load the CSV file on CPU
    input_filename = (
        f"/content/drive/MyDrive/Threat-Detection-in-Cyber-Security-Using-AI-master/"
        f"{attack_type}_vs_{benign_type}.csv"
    )
    attack_data = pd.read_csv(input_filename, low_memory=False)

    # Compute single-feature importances based on log_loss
    importances = perform_feature_selection_single_feature(attack_data)

    # Create a DataFrame for the results
    feature_names = attack_data.drop(columns=[" Label"]).columns
    importance_df = pd.DataFrame({
        "Feature": feature_names,
        "Importance": importances
    })

    # Percentage of total importance (this is optional)
    total_importance = importance_df["Importance"].sum()
    importance_df["Percentage"] = (
        importance_df["Importance"] / total_importance * 100
    )

    # Sort by importance descending
    importance_df = importance_df.sort_values(by="Importance", ascending=False)

    # Print the top 20 features
    print(f"\nTop 20 single-feature importances for {attack_type}:")
    print(importance_df.head(20))

    # Save the importance list to a CSV file
    importance_filename = f"{attack_type}_importance.csv"
    importance_df.to_csv(importance_filename, index=False)
    print(f"Saved single-feature importance list for {attack_type} -> {importance_filename}")

    # Plot the top 20 features
    top_20_df = importance_df.head(20)
    plt.figure(figsize=(10, 6))
    plt.bar(top_20_df["Feature"], top_20_df["Importance"], color="skyblue")
    plt.title(f"Single-Feature Importance (1/log_loss) for {attack_type}")
    plt.xlabel("Feature")
    plt.ylabel("Importance (higher is better)")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

print("Single-feature (log_loss-based) feature selection completed for all attack types.")
```
